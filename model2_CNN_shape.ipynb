```python
import os
import shutil

def process_images(source_directory, result_directory):
    # 결과 디렉토리가 없으면 생성
    if not os.path.exists(result_directory):
        os.makedirs(result_directory)

    # 파일을 순회하면서 이미지 처리
    for filename in os.listdir(source_directory):
        source_path = os.path.join(source_directory, filename)

        # shape1에 해당하는 알약 이미지
        if filename in shape1:
            result_path = os.path.join(result_directory, 'shape1')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape1'): {filename}")

        # shape2에 해당하는 알약 이미지
        elif filename in shape2:
            result_path = os.path.join(result_directory, 'shape2')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape2'): {filename}")
        # shape3에 해당하는 알약 이미지
        elif filename in shape3:
            result_path = os.path.join(result_directory, 'shape3')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape3'): {filename}")
        # shape4에 해당하는 알약 이미지
        elif filename in shape4:
            result_path = os.path.join(result_directory, 'shape4')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape4'): {filename}")
        # shape5에 해당하는 알약 이미지
        elif filename in shape5:
            result_path = os.path.join(result_directory, 'shape5')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape5'): {filename}")
        # shape6에 해당하는 알약 이미지
        elif filename in shape6:
            result_path = os.path.join(result_directory, 'shape6')
            process_images_in_directory(source_path, result_path)
            print(f"이미지 복사 ('shape6'): {filename}")

def process_images_in_directory(source_directory, destination_directory):
    # 결과 디렉토리가 없으면 생성
    if not os.path.exists(destination_directory):
        os.makedirs(destination_directory)

    # 이미지 파일을 순회하면서 복사하여 붙여넣기
    for filename in os.listdir(source_directory):
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            source_image_path = os.path.join(source_directory, filename)
            destination_image_path = os.path.join(destination_directory, filename)
            shutil.copyfile(source_image_path, destination_image_path)

if __name__ == "__main__":
    # 처리할 파일이 있는 디렉토리와 결과를 저장할 디렉토리 설정
    source_directory = "/content/drive/Othercomputers/내 노트북/졸업프로젝트/VS_1_단일"
    result_directory = "/content/drive/MyDrive/졸프/data/shape"

    # 이미지 처리 함수 호출
    process_images(source_directory, result_directory)
```

```python
import shutil
import random

def split( img_list, test_count, train_path, test_path):

  test_files=[]
  for i in random.sample( img_list, test_count ):
    test_files.append(i)

  # 차집합으로 train/test 리스트 생성하기
  train_files = [x for x in img_list if x not in test_files]

  for k in train_files:
    shutil.copy(k, train_path)

  for c in test_files:
    shutil.copy(c, test_path)

  print('train 폴더 이미지 개수 : {}\ntest 폴더 이미지 개수 : {}'.format(len(glob.glob(train_path+'/*')),len(glob.glob(test_path+'/*'))))
```

```python
def get_mean_std(dataset):
  meanRGB = [np.mean(image.numpy(), axis=(1,2)) for image,_ in dataset]
  stdRGB = [np.std(image.numpy(), axis=(1,2)) for image,_ in dataset]

  meanR = np.mean([m[0] for m in meanRGB])
  meanG = np.mean([m[1] for m in meanRGB])
  meanB = np.mean([m[2] for m in meanRGB])

  stdR = np.mean([s[0] for s in stdRGB])
  stdG = np.mean([s[1] for s in stdRGB])
  stdB = np.mean([s[2] for s in stdRGB])

  print(meanR, meanG, meanB)
  print(stdR, stdG, stdB)
```

## CNN 구현



```python
import torch
import torch.nn as nn # 신경망들이 포함됨
import torch.optim as optim # 최적화 알고리즘들이 포함됨
import torch.nn.init as init # 텐서에 초기값을 줌

import torchvision.datasets as datasets # 이미지 데이터셋 집합체
import torchvision.transforms as transforms # 이미지 변환 툴

from torch.utils.data import DataLoader # 학습 및 배치로 모델에 넣어주기 위한 툴

import numpy as np
import matplotlib.pyplot as plt
```

```python
batch_size = 50
learning_rate = 0.001
num_epoch = 4
```

```python
# Data preparation
# 이미지 크기 변환 후 텐서로 변환하는 전처리
transf = transforms.Compose([
    transforms.Resize((128, 128)),
    transforms.CenterCrop(128),
    transforms.ToTensor(),
    transforms.Normalize(resize_train_mean, resize_train_std),
])

trainset = torchvision.datasets.ImageFolder(root=train_path, transform=transf)
testset = torchvision.datasets.ImageFolder(root=test_path, transform=transf)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=batch_size,
                                          shuffle=True, num_workers=0)

testloader = torch.utils.data.DataLoader(testset, batch_size=batch_size,
                                         shuffle=False, num_workers=0)
```

```python
import torch.nn as nn
import torch.nn.functional as F

class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()

        # Convolutional layers
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(128, 256, kernel_size=3, padding=1)

        # Max pooling layer
        self.pool = nn.MaxPool2d(kernel_size=2, stride=2)

        # Fully connected layers
        self.fc1 = nn.Linear(256 * 16 * 16, 512)  # Adjust the input size based on your image size
        self.fc2 = nn.Linear(512, 6)  # Adjust the output size based on your task (e.g., number of classes)

    def forward(self, x):
        # Convolutional layers with ReLU activation and max pooling
        x = F.relu(self.conv1(x))
        x = self.pool(x)
        x = F.relu(self.conv2(x))
        x = self.pool(x)
        x = F.relu(self.conv3(x))
        x = self.pool(x)

        # Flatten before fully connected layers
        x = x.view(-1, 256 * 16 * 16)  # Adjust the size based on your image size

        # Fully connected layers with ReLU activation
        x = F.relu(self.fc1(x))
        #print(x)
        x = self.fc2(x)
        #print(x)

        x = F.log_softmax(x, dim=1)

        return x
```

```python
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
model = CNN().to(device)
```

```python
loss_func = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
loss_arr =[]
for i in range(num_epoch):
    for j,[image,label] in enumerate(trainloader):
        x = image.to(device)
        y= label.to(device)

        optimizer.zero_grad()

        output = model.forward(x)

        loss = loss_func(output,y)
        loss.backward()
        optimizer.step()

        if j % 20 == 0:
            print(loss)
            loss_arr.append(loss.cpu().detach().numpy())
```

```python
correct = 0
total = 0

# evaluate model
model.eval()

with torch.no_grad():
    for image,label in testloader:
        x = image.to(device)
        y= label.to(device)

        output = model.forward(x)

        # torch.max함수는 (최댓값,index)를 반환
        _,output_index = torch.max(output,1)

        # 전체 개수 += 라벨의 개수
        total += label.size(0)

        # 도출한 모델의 index와 라벨이 일치하면 correct에 개수 추가
        correct += (output_index == y).sum().float()

    # 정확도 도출
    print("Accuracy of Test Data: {}%".format(100*correct/total))
```
